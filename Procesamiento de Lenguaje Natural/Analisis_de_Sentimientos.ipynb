# **Análisis de sentimientos con NLTK**
Este algoritmo usa el ```vader_lexicon``` el cual contiene un conjunto de palabras ya puntuadas para entrenar el modelo y obtener un score sin necesidad de hacer un train previo para las categorias. El parametro ```compound``` entrega un resultado entre [-1,1] es decir normaliza el puntaje.

Para aplicarlo a Power BI no usamos las lineas

```df = pd.read_excel('comentarios.xlsx',sheet_name='Hoja2')```

dado que el conjunto de datos será el dataset en PowerQuery

```df```

Esta linea tampoco pues no requerimos imprimir el dataframe, y en lugar de ```df``` debemos usar el nombra para el conjunto de datos ```dataset```

````python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import nltk
nltk.download('vader_lexicon')

df = pd.read_excel('comentarios.xlsx',sheet_name='Hoja2')

cl = SentimentIntensityAnalyzer()

df['nltk_model'] = df['Comment'].apply(lambda x: cl.polarity_scores(x)['compound'])
df
````
